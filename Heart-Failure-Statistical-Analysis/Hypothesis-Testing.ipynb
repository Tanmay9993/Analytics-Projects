# Hypothesis Testing

# Chi-Squared Tests

# 1. Association Between Chronic Kidney Disease and 28-Day Mortality

**What We Are Doing:**

Testing whether there is a significant association between the presence of Chronic Kidney Disease (CKD) and 28-day mortality.

**Why We Are Doing It:**

* CKD can complicate heart failure and may increase the risk of mortality.
* Understanding this association can help in risk stratification and management.

**Aim:**

To determine if CKD significantly affects mortality rates in patients with heart failure.

**Hypothesis Formulation**
* Null Hypothesis (H₀): There is no association between CKD status and 28-day mortality.

* Alternative Hypothesis (H₁): There is an association between CKD status and 28-day mortality.

* Significance Level (α): 0.05

**Statistical Test:** Chi-Square Test of Independence

```python

# Create a contingency table
contingency_table = pd.crosstab(
    data['moderate.to.severe.chronic.kidney.disease'],
    data['death.within.28.days']
)

print("Contingency Table:")
print(contingency_table)

# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-Square Test Results:")
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the result
alpha = 0.05
if p < alpha:
    print("\nConclusion:")
    print("Reject the null hypothesis.")
    print("There is a significant association between CKD status and 28-day mortality.")
else:
    print("\nConclusion:")
    print("Fail to reject the null hypothesis.")
    print("There is no significant association between CKD status and 28-day mortality.")
```

# ANOVA Tests

# 1. Difference in Length of Stay Among Different Discharge Departments

**What We Are Doing:**

Testing whether there is a significant difference in Length of Stay (LOS) among patients discharged from different departments.

**Why We Are Doing It:**

* LOS may vary by department due to differences in care practices.
* Understanding this can help in resource planning and process improvement.

**Aim:**

To determine if LOS differs among departments.

**Hypothesis Formulation**
* Null Hypothesis (H₀): There is no significant difference in LOS among different departments.

* Alternative Hypothesis (H₁): At least one department has a different LOS.

* Significance Level (α): 0.05

**Statistical Test:** One-Way ANOVA

```python

# Prepare data
groups = data[['discharge.department', 'dischargeDay']].dropna()

# Get unique departments
departments = groups['discharge.department'].unique()

# Create a list of LOS for each department
los_groups = [groups[groups['discharge.department'] == dept]['dischargeDay'] for dept in departments]

# Ensure each group has at least 2 observations
los_groups = [group for group in los_groups if len(group) >= 2]

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(*los_groups)

print("One-Way ANOVA Results:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion:")
    print("Reject the null hypothesis.")
    print("There is a significant difference in LOS among different departments.")
else:
    print("\nConclusion:")
    print("Fail to reject the null hypothesis.")
    print("There is no significant difference in LOS among different departments.")
```

# 2. Difference in BMI Among Patients Receiving Different Medications

**What We Are Doing:**

Testing whether there is a significant difference in BMI among patients receiving different classes of medications.

**Why We Are Doing It:**

* Certain medications may be prescribed more frequently to patients with specific BMI profiles due to efficacy or safety considerations.
* Understanding this can help in personalizing treatment plans and managing medication-related risks.

**Aim:**

To determine if BMI differs significantly among patients receiving different medication classes.

**Hypothesis Formulation**
* Null Hypothesis (H₀): There is no significant difference in BMI among patients receiving different medication classes.

* Alternative Hypothesis (H₁): At least one medication class is associated with a significantly different BMI compared to others.

* Significance Level (α): 0.05

**Statistical Test:** One-Way ANOVA

```python

# Merge medications with patient BMI data
meds_classes = meds[['inpatient.number', 'Drug_name']].drop_duplicates()

# For patients with multiple medications, we'll take the first one (or you can define a priority)
meds_classes = meds_classes.groupby('inpatient.number').first().reset_index()

# Merge with BMI data
data_med = pd.merge(
    data[['inpatient.number', 'BMI']],
    meds_classes,
    on='inpatient.number',
    how='inner'
)

# Ensure 'BMI' is numeric
data_med['BMI'] = pd.to_numeric(data_med['BMI'], errors='coerce')

# Prepare data
groups = data_med[['Drug_name', 'BMI']].dropna()

# Get unique medication classes
med_classes = groups['Drug_name'].unique()

# Create a list of BMI values for each medication class
bmi_groups = [groups[groups['Drug_name'] == med]['BMI'] for med in med_classes]

# Ensure each group has at least 2 observations
bmi_groups = [group for group in bmi_groups if len(group) >= 2]
valid_classes = [med_classes[i] for i in range(len(med_classes)) if len(bmi_groups[i]) >= 2]

# Perform One-Way ANOVA
if len(bmi_groups) >= 2:
    f_stat, p_value = stats.f_oneway(*bmi_groups)
    print("One-Way ANOVA Results:")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("\nConclusion:")
        print("Reject the null hypothesis.")
        print("There is a significant difference in BMI among patients receiving different medication classes.")
    else:
        print("\nConclusion:")
        print("Fail to reject the null hypothesis.")
        print("There is no significant difference in BMI among patients receiving different medication classes.")
else:
    print("Not enough groups with sufficient data for ANOVA.")
```
